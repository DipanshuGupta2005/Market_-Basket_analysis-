```python
# %pip install sqlalchemy psycopg2-binary --quiet

from sqlalchemy import create_engine
import psycopg2
import os
import sqlalchemy
import time
from sqlalchemy import exc as sa_exc

# Ensure DataFrame 'df' exists (load from CSV if missing)
if 'df' not in globals():
	# Try the common filenames used in this notebook
	for fname in ("shopping_behavior_updated.csv", "customer_behavior_updated.csv"):
		if os.path.exists(fname):
			df = pd.read_csv(fname)
			print(f"Loaded DataFrame 'df' from '{fname}'.")
			break
	else:
		# If neither file exists, raise a clear error
		try:
			# final attempt using the primary expected filename
			df = pd.read_csv("shopping_behavior_updated.csv")
			print("Loaded DataFrame 'df' from 'shopping_behavior_updated.csv'.")
		except Exception as e:
			raise NameError(
				"DataFrame 'df' is not defined and could not be loaded. "
				"Expected CSV files: 'shopping_behavior_updated.csv' or 'customer_behavior_updated.csv'. "
				"Ensure the file is present in the working directory."
			) from e

# Create engine only if missing to avoid overwriting an existing connection
if 'engine' not in globals():
	username = globals().get('username', 'postgres')
	password = globals().get('password', '1234')
	host = globals().get('host', 'localhost')
	port = globals().get('port', '5432')
	database = globals().get('database', 'customer_behaviour')

	conn_url = f'postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}'

	def _create_engine_and_connect(url):
		engine = create_engine(url)
		# try a short-lived connection to validate
		with engine.connect() as conn:
			pass
		return engine

	try:
		# create the engine and test a connection immediately to surface errors
		engine = _create_engine_and_connect(conn_url)
	except Exception as e:
		err_msg = str(e).lower()
		# If the error indicates the database does not exist, attempt to create it
		if 'does not exist' in err_msg or ('database' in err_msg and 'does not exist' in err_msg):
			print(f"Database '{database}' not found. Attempting to create it...")
			try:
				# connect to the default 'postgres' database to create the target database
				conn_params = dict(dbname='postgres', user=username, password=password, host=host, port=port)
				# establish a raw connection and set autocommit/isolation level so CREATE DATABASE can run
				create_conn = psycopg2.connect(**conn_params)
				try:
					# Ensure CREATE DATABASE is executed outside a transaction block.
					# Prefer using ISOLATION_LEVEL_AUTOCOMMIT; fall back to .autocommit if needed.
					try:
						from psycopg2.extensions import ISOLATION_LEVEL_AUTOCOMMIT
						create_conn.set_isolation_level(ISOLATION_LEVEL_AUTOCOMMIT)
					except Exception:
						create_conn.autocommit = True

					with create_conn.cursor() as cur:
						cur.execute("SELECT 1 FROM pg_database WHERE datname = %s", (database,))
						if not cur.fetchone():
							cur.execute(f'CREATE DATABASE "{database}"')
							print(f"Database '{database}' created.")
						else:
							print(f"Database '{database}' already exists (race condition).")
				finally:
					create_conn.close()

				# small pause to allow the server to register the new DB
				time.sleep(1)
				# recreate the SQLAlchemy engine now that the DB should exist
				engine = _create_engine_and_connect(conn_url)
			except Exception as create_err:
				# surface helpful message for common causes
				msg = str(create_err).lower()
				if 'could not connect to server' in msg or 'connection refused' in msg:
					raise RuntimeError(
						f"Could not connect to Postgres server at {host}:{port}. Ensure the server is running and credentials are correct."
					) from create_err
				else:
					raise RuntimeError(f"Failed to create database '{database}': {create_err}") from create_err
		elif 'could not connect to server' in err_msg or 'connection refused' in err_msg:
			# server isn't reachable
			raise RuntimeError(
				f"Could not connect to Postgres server at {host}:{port}. Ensure the server is running and accessible."
			) from e
		else:
			# unknown error, re-raise for visibility
			raise

table_name = globals().get('table_name', 'customer')

# Upload dataframe
df.to_sql(table_name, engine, if_exists="replace", index=False)
print(f'data successfully loaded into table \"{table_name}\" in database \"{database}\".')
```
